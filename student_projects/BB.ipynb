```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
```

הצצה לדאטא מה משפיע על הדירוג העולמי של העסקים  

```python
Data=pd.read_csv('Rankings.csv')
data=Data.dropna()
first_10_countries=data.head(10)
last_10_countries = data.tail(10)
data.head(50)
```

(1) trading--global

רוב האנשים מחברות שונות בעולם חושבים שכדי שעסק יצליח הוא צריך לסחור גם ברמה הבינלאומית ולא רק ברמה המקומית, אך לפי הניתונים בעולם אנו רואים שתיאוריה זו לא בהכרח נכונה, יש הרבה עסקים מוצלחים מאוד והם מתעסקים רק במסחר המיקומי, וההיפך נכון.
למשל: ניו זילנד עומדת במקום הראשון בדירוג הגלובלי אך בדירוג נמוך במסחר מעבר לגבולות (במקום 63). 

```python
plt.title('how much the trading across borders important to global rank')
plt.xlabel('globalRank')
plt.ylabel('Trading across Borders')
plt.plot(data['globalRank'],data['Trading across Borders'])
plt.show()
```

(3) first 10 countries

```python
corr_df=first_10_countries[['Dealing with Construction Permits','globalRank']]
corr_df.corr()
```

(4)last 10 countries

לקחנו את המדינות עם הדירוג הכי גבוה במסחר הגלובלי ובדקנו מה הסיבה האמיתית לכך שהם במקומות האחרונים

```python

corr_df=last_10_countries[['Starting a Business','Dealing with Construction Permits']]
corr_df.corr()
```

```python
Corrs_df1 = corr_df.corr(method = 'pearson')
plt.figure(figsize=(5,5))  
sns.heatmap(Corrs_df1, cmap='twilight', center=1, annot = True);
```

ניסינו לבדוק מה הסיבה לכך שמדינות אלו במקומות האחרונים ולפי הקורולצייה גם כאן ההתמודדות עם היתרי הבניה במדינה הם אלה שמשפיעים על הדירוג במסחר הגלובלי של המדינה.

האם יש עוד סיבות משמעותיות שמשפיעות על הדירוג הגלובלי?

```python

corr_df1=data[['globalRank','Protecting Minority Investors','Resolving Insolvency']]
corr_df1.corr()
```

```python
Corrs_df2 = corr_df1.corr(method = 'pearson')
plt.figure(figsize=(5,5))  
sns.heatmap(Corrs_df2, cmap='twilight', center=1, annot = True);
```

לפי האיור ניתן לראות שההגנה על משקיעי מיעוטים היא גם משפיעה על הדירוג הגלובלי ניתן להבין שבמדינות העניות יחסית, אכיפת החוק במדינה חלשה, התנאים האלה יוצרים קושי בהגנה על משקיעי מיעוטים מה שגורם לעליית הדירוג הגלובלי, וההיפך נכון - המדינות במקומות הראשונים בדירוג הגלובלי הן לרוב מדינות עם תנאים יותר טובים. 

```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(data.corr()[['globalRank']].sort_values(by='globalRank', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Rating of influencing factors', fontdict={'fontsize':18}, pad=16);
```

לסיכום, הרבה אנשי עסקים מניחים תיאוריות לפני פתיחת העסק שלהם המתמקדות באיך לגרום לעסק להצליח ביותר, אבל חלק גדול מהתיאוריות האלו אינן בהכרח נכונות ואין נתונים מוכחים בעולם שמראות שיש קשר בין התיאוריות האלו להצלחת העסק, כמו תשלומי המיסים, ייצור ומסחר מחוץ לגבולות כמו שאנחנו חשבנו.
